## LocalTrust

```python
import requests

base_url = 'https://base-sepolia.easscan.org/attestations/forSchema/0x5ee00c7a6606190e090ea17749ec77fe23338387c23c0643c4251380f37eebc3?_data=routes%2F__boundary%2Fattestations%2FforSchema%2F%24schemaUID&limit=50'

def get_attestations(page=1):
    results = []

    def run(page):
        nonlocal results
        url = base_url
        if page > 1:
            url += f'&page={page}'

        res = requests.get(url).json()
        results.extend(res['attestations'])

        if len(results) < res['totalAttestationCount']:
            return run(page + 1)

        return results

    return run(page)

attestations = get_attestations()
localtrust = [{'i': r['attester'], 'j': r['recipient'], 'v': 1 } for r in attestations if r['attester'] != r['recipient']]
df = pandas.DataFrame(localtrust)
df
```

##EigenTrust

```python
scores = a.run_eigentrust(localtrust)
df2 = pandas.DataFrame(scores)
df2
```
